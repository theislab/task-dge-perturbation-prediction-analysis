```python
import scanpy as sc
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_dir = "/Users/arturszalata/PycharmProjects/task-dge-perturbation-prediction/resources/neurips-2023-raw"
reannotated_h5ad = f'{data_dir}/sc_counts_reannotated_with_counts.h5ad'
```

```python
adata = sc.read_h5ad(reannotated_h5ad)
```

```python
palette = sns.color_palette("colorblind", n_colors=adata.obs['cell_type'].nunique())

sc.pl.umap(adata, color='cell_type', palette=palette, show=False, legend_loc='none', title="")

ax = plt.gca()
ax.set_axis_off()

ax.figure.set_facecolor('none')
ax.set_facecolor('none')

plt.tight_layout()

plt.savefig("umap_cell_type_annotation.pdf", transparent=True)
```

```python
palette = ["#949494", "#ECE133", "#56B4E9"]
sc.pl.umap(adata, color='donor_id', palette=palette, show=False, legend_loc='none', title="")

ax = plt.gca()
ax.set_axis_off()

ax.figure.set_facecolor('none')
ax.set_facecolor('none')

plt.tight_layout()

plt.savefig("umap_donor.pdf", transparent=True)
```
