```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

np.random.seed(42)
palette = sns.color_palette("colorblind", n_colors=4)

data = {
    'Sample Type': ['Control'] * 3 + ['Drug A'] * 3 + ['Drug B'] * 3 + ['Drug C'] * 3,
    'Response': np.concatenate([
        np.random.normal(10, 1, 3),  # Control
        np.random.normal(10, 1, 3),  # Drug 1
        np.random.normal(13, 1, 3),  # Drug 2
        np.random.normal(15, 1, 3)   # Drug 3
    ])
}

df = pd.DataFrame(data)

plt.figure(figsize=(8, 6), frameon=False) 
ax = sns.stripplot(x=np.zeros(df.shape[0]), y='Response', hue='Sample Type', data=df, jitter=True, dodge=False, palette=palette, size=20)

ax.set(yticks=[])
ax.set_ylabel('')

ax.set(xticks=[])
ax.set_xlabel('')
ax.set(frame_on=False) 

for spine in ax.spines.values():
    spine.set_visible(False)

ax.patch.set_visible(False)
plt.gca().set_axis_off()

# legend = plt.legend(title='Sample Type', loc='upper right', frameon=False)
# hide legend
ax.get_legend().remove()

plt.savefig("stripplot_donor1.svg", bbox_inches='tight', transparent=True)
plt.show()
```
